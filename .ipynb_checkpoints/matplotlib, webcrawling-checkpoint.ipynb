```python
import matplotlib.pyplot as plt
import numpy as np

plt.rcParams['font.family'] = 'new Gulim'
plt.rcParams['font.size'] = 10

X = np.linspace(0, 2 * np.pi, 100)
Ya, Yb = np.sin(X), np.cos(X)

plt.plot(X, Ya)
plt.plot(X, Yb)

plt.show()
```

```python
X = np.linspace(0, 2 * np.pi, 100)
Ya, Yb = np.sin(X), np.cos(X)

plt.plot(X, Ya, label='sin(X)')
plt.plot(X, Yb, label='cos(X)')

plt.legend()
plt.grid(linestyle='--')

plt.show()
```

```python
import matplotlib.pyplot as plt
from PIL import Image 

im = Image.open('apple_logo.jpg')
im = im.resize((298, 373))

plt.imshow(im)
plt.show()
```

```python
from matplotlib import patches

plt.ylim(400, 0)
plt.xlim(0, 400)

x = [100, 100, 114, 146, 158, 160, 143, 112, 100]
y = [148, 138, 125, 127, 141, 156, 169, 168, 148]

plt.imshow(im)
line = plt.Line2D(x, y, color='red')

axes=plt.gca()
axes.add_line(line)

c = patches.Circle(xy=(50,300), radius=25, fill=False, facecolor='b', edgecolor='r')

r = patches.Rectangle(xy=(50, 50), width=10, height=25)
e = patches.Ellipse(xy=(100, 100), width=20, height=50, angle=60)

axes.add_patch(e)
axes.add_patch(r)

axes.add_patch(c)

plt.show()
```

```python
def download(url, file):
    response = requests.get(url)
    
    with open(file, "wb") as f:
        f.write(response.content)
        
def norm_file_name(file):
    bad_chars = '\\/:*?<>|'
    for c in bad_chars:
        file = file.replace(c, ' ')
    return file

driver = webdriver.Chrome('chromedriver.exe')
driver.get('https://movie.naver.com/movie/running/current.nhn')

a_movies = driver.find_elements_by_xpath("//div[@class='thumb']//img")

for idx, a_movie in enumerate(a_movies, 1):
    title = a_movie.get_attribute('alt')
    src = a_movie.get_attribute('src')
    
    download(src, 'thumb/' + norm_file_name(title) + '.jpg')
#    print(title)
#    print(src)
    
driver.quit()
```
