```python
f = open('file.txt', 'w', encoding='utf-8')
f.write('파일에 기록될 문자열입니다.\n')
f.write('파일에 기록하는거 참 쉽죠?!\n')
f.close()

f = open('file.txt', 'a', encoding='utf-8')
f.writelines(['하나', '둘', '셋', '넷', '다섯'])
f.write('\n')
f.writelines('\n'.join(['여섯', '일곱', '여덟', '아홉', '열']))
f.close()
```

```python
f = open('file.txt', 'r', encoding='utf-8')
s = f.read() #read : 파일에 있는 데이터를 한 번에 모두 불러옴
print(s)
f.close()
```

```python
f = open('file.txt', 'r', encoding='utf-8')
s = f.readline() #readline : 파일을 한 라인씩 읽어옴
print(s, end='')
for s in f:
    print(s, end='')
f.close()
```

```python
f = open('file.txt', 'r', encoding='utf-8')
lines = f.readlines() #readlines : 라인 별 끊어서 리스트에 저장
print(lines)
s = ''.join(lines)
print(s)
f.close()
```

```python
#파일 현재 위치 확인(tell) 및 변경(seek)
f = open('file.txt', 'r', encoding='utf-8')
print('파일 현재 위치:', f.tell())
print(f.read())
print('파일 현재 위치:', f.tell())

#처음으로
f.seek(0)
print('파일 현재 위치:', f.tell())
print(f.readline(), end='')
print('파일 현재 위치:', f.tell())
f.close()
```

```python
# with 블록을 벗어났을 때 자동으로 파일이 닫힘
with open('file.txt', 'r') as f:
    try:
        for s in f:
            print(s, end='')
    except:
        print('여기서 예외처리를 합니다.')
```

```python
# csv : 각 라인의 컬럼들이 컴마로 분리된 텍스트 파일 포맷. csv 특성 상 가장 윗 부분은 각 데이터의 설명 및 이름명이 존재 할 경우가 많으며 
# 실제 데이터 처리에는 필요 없는 부분이므로 next()함수를 실행하여 생략 가능

import csv

f = open('공유/data-iris-1.csv', 'r')
rdr = csv.reader(f)
for line in rdr:
    print(line)
f.close()
```

```python
import csv

with open('공유/covid_19_data.csv', 'r') as f:
    rdr = csv.reader(f)
    next(rdr)
    for line in rdr:
        print(line)
```

```python
# writerows() : 리스트에 미리 데이터를 넣어두고 한 번에 쓰기

import csv

with open('공유/write2.csv', 'a', newline='') as f:
    wr = csv.writer(f)
    wr.writerow([1, '대구', 6181])
    wr.writerow([2, '경북', 1316])
    wr.writerow([3, '경기', 580])
```

```python
# json : 데이터 교환 포맷. 파이썬에서 딕셔너리와 동일한 양식을 가지고 있어 쉽게 사용 및 가공이 가능

import json

dog = {"name":"식빵","family":"웰시코기","age":1,"weight":2.14}

json_string = json.dumps(dog, indent=4, ensure_ascii=False)
print(json_string)
print(type(json_string))

json_dict = json.loads(json_string)
print(json_dict)
print(type(json_dict))
```

```python
import json

with open('공유/covid_19_data_json.json', 'r') as f:
    cov_list = json.load(f)
    for cov_info in cov_list:
        print("="*25)
        print(cov_info)
```

```python
import json

with open('공유/cov_count.json', 'w', encoding='utf-8') as f:
    cov_info = [{"id":1, "region": "대구", "count": 6781}, {"id":2, "region": "경북", "count": 1316}, 
                {"id":3, "region": "경기", "count": 580} ]
    json.dump(cov_info, f, ensure_ascii=False, indent=4, sort_keys=True)
```

```python
from collections import OrderedDict

keys = ["id","region", "count"]
value_list = [[1, "대구, 6781"], [2,"경북", 1316], [3,"경기",580]]
cov_info = [OrderedDict((key,value) for key, value in zip(keys, values)) for values in value_list]
print(cov_info)
```

```python
from PIL import Image, ImageDraw, ImageEnhance

im = Image.open('apple_logo.jpg')

print(im.filename)
print(im.size)
print(im.width, im.height)
print(im.format)

im.save('im_save.jpg')
im.show()
```

```python
# 잘라낼 영역은 좌표 (x1, y1, x2, y2) 좌표 원점(0, 0)은 이미지의 좌측 상단

area = (200, 100, 2500, 2500)
im_crop = im.crop(area)
im_crop.show()
```

```python
# 이미지 생성시 색상모드, 크기(w, h), 배경색상(RGB) 전달, 붙여 넣을때는 목적지 이미지의 좌표를 전달

im_paste = Image.new("RGB", size=(600,700), color=(100,200,300))
im_paste.paste(im, (0,0))
im_paste.show()
```

```python
drawer = ImageDraw.Draw(im)
drawer.line(xy=((10,10),(1000,1000)),fill='BLUE', width=10)
drawer.point(xy=((150,150),(250,250),(350,350)),fill='GREEN')
drawer.polygon(xy=((450,450),(550,550),(650,650)),fill='PINK')
drawer.rectangle(xy=((750,750),(850,850)),fill='YELLOW')

im.show()
```

```python
enhancer = ImageEnhance.Color(im)
im_gray = enhancer.enhance(1.0)
im_gray.show()
```

```python
# 이미지 회전

im_rot = im.rotate(20, expand=False)
im_rot.show()
```

```python
# 이미지 크기 조절

im_res = im.resize((im.width // 2, im.height // 2))
im_res.show()
```

```python
# 이미지 반전

flip_lr = im.transpose(Image.FLIP_LEFT_RIGHT)
flip_tb = im.transpose(Image.FLIP_TOP_BOTTOM)

flip_lr.show()
flip_tb.show()
```

```python
m = [1.5, -0.5, 0.5, 0.0, 1.0, 0.0]
im_affine = im.transform(size=(im.width, im.height),method=Image.AFFINE,data=m)
im_affine.show()
```
